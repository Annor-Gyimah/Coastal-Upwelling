```python
def ensure_time_dim(da):
    """
    Ensures DataArray has a single time dimension named 'time'.
    Safely converts 'valid_time' → 'time' and removes duplicates if needed.
    """
    # Case 1: already good
    if 'time' in da.dims:
        return da

    # Case 2: 'valid_time' is a dimension, but there's also a 'time' coordinate → drop and rename
    if 'valid_time' in da.dims and 'time' in da.coords:
        da = da.drop_vars('time')
        da = da.rename({'valid_time': 'time'})
        return da

    # Case 3: 'valid_time' is a coordinate, not a dimension
    if 'valid_time' in da.coords and 'valid_time' not in da.dims:
        da = da.swap_dims({'valid_time': 'time'})
        return da

    # Case 4: neither time-like coordinate found
    raise ValueError("No recognizable time dimension ('time' or 'valid_time') found.")
```

```python
def fix_valid_time_conflict(da):
    """
    Fixes datasets where 'valid_time' is the dimension and 'time' is a coordinate.
    Converts 'valid_time' → 'time' safely.
    """
    if 'valid_time' in da.dims:
        if 'time' in da.coords:
            da = da.drop_vars('time')  # drop conflicting coordinate
        da = da.rename({'valid_time': 'time'})
    return da
```
